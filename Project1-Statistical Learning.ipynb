```python
#importing numpy and pandas and checking the working directory

import numpy as np
import pandas as pd
import math
import os
os.getcwd()
```

```python
# making a pandas dataframe la_liga_df by reading Laliga.csv into it with header

la_liga_df = pd.read_csv("Laliga.csv",header=1)
la_liga_df.head(10)
```

```python
#Replacing dashes with Zero in the dataframe la_liga_df

la_liga_df = la_liga_df.replace(to_replace='-',value=0)
la_liga_df.head(10)
```

```python
#Teams which have started playing between 1930-1980 in la_liga

la_liga_debut = la_liga_df.loc[(la_liga_df['Debut'] >= '1930') & (la_liga_df['Debut'] <= '1980')]
la_liga_debut[['Team','Debut']]
```

```python
#List of teams which came Top 5 in terms of points in la_liga

la_liga_df['Points'] = la_liga_df['Points'].astype('int')
la_liga_top_five = la_liga_df.nlargest(5,'Points')
la_liga_top_five[['Team','Points']]
```

```python
#Function with name â€œGoal_diff_countâ€ to calculate the goal difference of each team in la_liga

def Goal_diff_count (x):
    laliga = pd.read_csv(x,header=1)
    
    team_goal_diff = laliga[['Team','GoalsFor','GoalsAgainst']]
    
    team_goal_diff = team_goal_diff.replace(to_replace='-',value=0)
    
    team_goal_diff[['GoalsFor','GoalsAgainst']] = team_goal_diff[['GoalsFor','GoalsAgainst']].astype('int')
    
    team_goal_diff['Goal Differences'] = team_goal_diff['GoalsFor'] - team_goal_diff['GoalsAgainst']
    
    print ("The maximum goal difference is {}".format((team_goal_diff['Goal Differences'].max())))
    print ("The minimum goal difference is {}".format((team_goal_diff['Goal Differences'].min())))
    
    return team_goal_diff
    
Goal_diff_count('Laliga.csv')  
```

```python
#Calculate the â€œWinning Percentâ€ for all the teams.

la_liga_df['GamesWon'] = la_liga_df['GamesWon'].astype('int')
la_liga_df['GamesPlayed'] = la_liga_df['GamesPlayed'].astype('int')
la_liga_df['Winning Percent'] = (la_liga_df['GamesWon']/la_liga_df['GamesPlayed']) * 100
la_liga_df[['GamesWon','GamesPlayed','Winning Percent']]
```

```python
#Taking the top five in terms of win percentage by replacing the Nan values with zero first in the dataframe la_liga_df

la_liga_df['Winning Percent'].fillna(0, inplace=True)
la_liga_top_five_win_percent = la_liga_df.nlargest(5,'Winning Percent')
la_liga_top_five_win_percent[['Team','Winning Percent']]
```

```python
#Teams grouped on their â€œBest positionâ€ and their respective sum of points for each position

la_liga_best = la_liga_df.groupby('BestPosition')['Points'].sum()
print (la_liga_best)
```
